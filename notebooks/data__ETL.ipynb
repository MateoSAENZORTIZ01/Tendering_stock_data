```python
import requests
import time

RESOURCE_ID = "814b7b54-477a-4c25-b3bf-6be05412069d"
BASE_URL = "https://data.iadb.org/api/action/datastore_search"

def fetch_all_projects(batch=100):
    offset = 0
    all_records = []

    while True:
        params = {
            "resource_id": RESOURCE_ID,
            "limit": batch,
            "offset": offset
        }
        
        r = requests.get(BASE_URL, params=params).json()
        records = r["result"]["records"]
        all_records.extend(records)
        
        print(f"Fetched {len(records)} records (offset={offset})")
        
        if len(records) < batch:
            break
        
        offset += batch
        time.sleep(0.2)  # be nice to the server

    return all_records

projects = fetch_all_projects()

print("TOTAL PROJECTS DOWNLOADED:", len(projects))
```
